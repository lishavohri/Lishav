```python
import pandas as pd
import numpy as np

udemy=pd.read_csv('udemy_output_All_Finance__Accounting_p1_p626 (1).csv')
udemy
udemy.head(2)
```

```python
# To know the total number of rows and columns
total_rows=len(udemy.axes[0])
total_cols=len(udemy.axes[1])
print("Number of Rows: "+str(total_rows))
print("Number of Columns: "+str(total_cols))
```

```python
udemy.drop(['discount_price__price_string', 'price_detail__price_string'], axis=1,inplace=True)
udemy.head(2)
```

```python
# 1. To differenciate number of uploaded course each year
import matplotlib.pyplot as plt
udemy['year'] = pd.DatetimeIndex(udemy['published_time']).year
udemy.head()
x=udemy.groupby("year")["url"].count()
x.plot(kind="bar", figsize=(6,6))
plt.title('Number of uploaded course each year')
plt.xlabel('year')
plt.ylabel('Total number of course')
```

```python
# 2. Any one wants to know course-related Management in Udemy.

Management=udemy[udemy.url.str.contains('pmp'or'mba'or'agile'or'product manager'or'leadership'or
                                        'foundation-course'or'project-management'or'process-mapping'or
                                        'product-management'or'scrum',case=False)]
Management.head(5)
```

```python
# 2.1.To find the Price range for Management course

Management.is_paid.unique()
Management.discount_price__amount.unique()
```

```python
#3. To look at a rating for course-related "start_up"
start_up=udemy[udemy.url.str.contains('start-up',case=False)]
start_up
start_up.groupby("rating")["id"].count()
```

```python
# 4 To find all course-related finance with high to low subscriber.
Finance=udemy[udemy.url.str.contains('financial-analyst' or 'investment-banking'or'forex-trading'or
                                     'accounting'or'trading'or'finance'or'trader'or'financial-analysis'or
                                     'financial-planning' or 'stock-market',case=False)]
Finance.sort_values(by="num_subscribers",ascending=False)
 
```

```python
# 5.To know the percentage rate of courses uploaded related to IT compare to all
IT=udemy[udemy.url.str.contains('business-analysis'or'business intelligence'or'hadoop'or
                                'numpy'or'python'or'data-analysis'or'data-analytics',case=False)]
IT.head()
Per_IT_Course=IT['title'].count()/udemy['title'].count()*100
Per_IT_Course
```

```python
# 6. Show the total paid and free course available
udemy.groupby("is_paid")["id"].count()
```

```python
# 7. Convert data of title and number of lectures to the dictionary
description_dict = udemy.set_index('title')['num_published_lectures'].to_dict()
description_dict
```

```python
# 8. Check the distrubiton of rating
udemycolumns = ['avg_rating','avg_rating_recent','rating']
udemy[udemycolumns].hist(bins=20,figsize=(10,11), color='blue', edgecolor='yellow',linewidth=2);
```

```python
# 9. Show the rating wise lecture
udemy.groupby("rating")["id"].count()
```

```python
# 10. Top 3 course as per the rating
udemy.dropna(inplace = True) 
top3 = udemy.nlargest(3, "rating") 
top3
```

```python
# 11. Course title having maximum price
udemy[udemy['price_detail__amount'] == udemy['price_detail__amount'].max()]
```

```python
# 12. Statical functions on data
mean1 = udemy['price_detail__amount'].mean()
sum1 = udemy['price_detail__amount'].sum()
max1 = udemy['price_detail__amount'].max()
min1 = udemy['price_detail__amount'].min()
count1 = udemy['price_detail__amount'].count()
median1 = udemy['price_detail__amount'].median() 
print ('Mean of price_detail__amount: ' + str(mean1))
print ('Sum of price_detail__amount: ' + str(sum1))
print ('Max price_detail__amount: ' + str(max1))
print ('Min price_detail__amount: ' + str(min1))
print ('Count of price_detail__amount: ' + str(count1))
print ('Median price_detail__amount: ' + str(median1))
```
